
***PubMed PICO Element Detection Dataset***

- Test Set: 2400 abstract - 2723 frasi etichettate come "Participants" (P), 2331 come "Intervention" (I) e 3219 come "outcome" (O);
- Train Set: 19868 abstract -  22347 frasi etichettate come "Participants" (P), 19864 come "Intervention" (I) e 26229 come "outcome"(O);
- Dev Set: 2400 abstract - 2625 frasi etichettate come "Participants" (P), 2407 come "Intervention" (I) e 3077 come "outcome" (O).

  Gli abstract sono stati estratti da PubMed assecondando i seguenti limiti:
- Text Availability: Abstract; 
- Languages: English; 
- Types: Randomized Controlled Trial

Il processo automatico di etichettatura dei dati è stato supportato dalla presenza nei testi di determinate keywords associate alle sette labels stabilite dai ricercatori.  

- Aim (A) --> Objective, Background, Purpose, Importance, Introduction, Aim, Rationale, Goal, Context, Hypothesis;
- Participants (P) --> Population, Participant, Sample, Subject, Patient;
- Intervention (I) --> Intervention;
- Outcome (O) --> Outcome, Measure, Variable, Assessment;
- Method (M) --> Method, Setting, Design, Material, Procedure, Process,Methodology;
- Results (R) --> Result, Finding;
- Conclusion (C) --> Conclusion, Implication, Discussion, Interpretatio.

L'etichetta associata a Comparison (lettera 'C' nell'acronimo PICO) è stata assimilata all'interno della label 'I' data la sovrapposizione delle due nella tipologia di intervento chiamata 'Randomized Controlled Trial'.

*Link alla risorsa* --> https://github.com/jind11/PubMed-PICO-Detection

***Nicta_Piboso Dataset***

- Test Set: 200 abstract - 69 frasi etichettate come "Participants" (P), 61 come "Intervention" (I) e 948 come "outcome" (O);

- Train Set: 799 abstract - 331 frasi etichettate come "Participants" (P), 279 come "Intervention" (I) e 3504 come "outcome"(O).

I dati sono stati selezionati da una lista di interrogazioni tratte da  *Global Evidence Mapping Initiative (GEM)* e da Agency for *Healthcare Research and Quality (AHRQ)*. Il processo di annotazione è stato condotto manualmente da studenti di medicina supportati da adeguati supervisori. L'assegnazione delle labels è avvenuta tramite interannotator agreement basato su Cohen’s Kappa values. 

- Background (B)
- Population (P)
- Intervention (I)
- Outcome (O)
- Study (S)
- Other (OT)

````python
import pandas as pd
import random
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
from collections import defaultdict
import os
from google.colab import drive,files


def data_preparation (abstract):
  
  for index,text in enumerate(abstract[abstract_columns[0]].tolist()[1:]):
      sentence = list(sent_tokenize(text))
      labels = ["A", "M", "P", "I", "O", "R", "C"]
      random_name = ['AIM','BACKGROUND','DESIGN','PATIENTS','CONCLUSION','RESULTS']
      sentence_list = [f"{random.choice(random_name)}|{random.choice(labels)}|{sent}" for i,sent in enumerate(sentence)]
      sentence_txt = "###"+str(index)+":\n"+ '\n'.join(sentence_list) +'\n\n' 
  
      with open('/content/drive/MyDrive/CNR/Deep-PICO-Detection/data/pico/prove/test.txt','a',encoding="utf-8") as out:
            out.write(sentence_txt)
  print()
  print("*******************************")
  print('Data ready for classification.')  


def data_processor (eval_text,test):

  with open(eval_text,'r',encoding="utf-8") as doc:
      with open(test,'r',encoding="utf-8") as test:
          sentence = [ele.split('\t') for ele in doc.read().split('\n')]
          sentence = [ele[1:] for ele in sentence]
          test_sent = test.read().split('###')
                                                          
          res = [(index,ab.index(sent),label,sent) for label,sent in sentence[:-1] 
                  for index,ab in enumerate(test_sent) if sent in ab and sent != ""]

          diz = {k:[(pos,sent,label) for i,pos,label,sent in res if i == k] for k in [index[0] for index in res]}        
          diz = dict(sorted(diz.items()))
         
          diz_ = defaultdict(list)
          for k,v in diz.items():
              for ele in v:
                  if ([ele[0] for ele in v]).count(ele[0]) == 1:
                      diz_[k].append(ele)
                  else:
                      if len(ele[1].strip().split()) > 2 and ele[1] not in [ele[1] for ele in diz_[k]]:
                          diz_[k].append(ele)
                      
          
          diz_ = {k:sorted(v, key=lambda tup: tup[0]) for k,v in diz_.items()}
                    
          txt = {k: ' '.join([f"{sent[:-1]} [{label}]." for pos,sent,label in v]) for k,v in diz_.items()}               
          txt = [f"###{k-1}\n {v}\n\n" for k,v in txt.items()]
                          
          out = pd.DataFrame(txt)
          #file_csv          
          out.to_csv('output_abstract.csv',index=False)
          print()
          print("*************************")
          print('Classification performed.')



````
